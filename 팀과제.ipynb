(1) 카테고리별 좋아요/싫어요 비율 분포 알아보기

```python

import json

pd.read_json('C:/Temp/KR_category_id.json')

id_to_category = {}

with open('C:/Temp/KR_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        id_to_category[category['id']] = category['snippet']['title']

        
id_to_category
```

```python
KRvideos2 = KRvideos.groupby([KRvideos.category_id], as_index = False).mean()
```

```python
KRvideos2["Rate"] = KRvideos2["likes"] / KRvideos2["dislikes"]
```

```python
KRvideos2.dropna(axis=0)
```

```python
KRvideos2.sort_values(by=["Rate"], ascending = False)
```

```python
JPvideos2 = JPvideos.groupby([JPvideos.category_id], as_index = False).mean()
JPvideos2["Rate"] = JPvideos2["likes"] / JPvideos2["dislikes"]
JPvideos2.dropna(axis=0)
JPvideos2.sort_values(by=["Rate"], ascending = False)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns


```

```python
ax = plt.subplots()
ax = sns.barplot(x = 'category_id', y = 'Rate', data = KRvideos2)
```

```python
ax = plt.subplots()
ax = sns.barplot(x = 'category_id', y = 'Rate', data = JPvideos2)
```

한국과 일본 모두 10번 카테고리, 즉 음악 부분이 가장 평균 좋아요 비율이 높은 것을 알 수 있다. 
하지만 가장 적은 평균 좋아요 비율을 기록한 카테고리는 서로 달랐는데, 한국은 '공연' 카테고리가, 일본은 '여행' 카테고리가 가장 낮았다. 

#(2) 한국과 일본의 평균 년도별 유튜브 조회수 알아보기

```python
KRvideos['time'] = pd.to_datetime(KRvideos['publish_time'])
JPvideos['time'] = pd.to_datetime(JPvideos['publish_time'])

KRvideos['Year'] = KRvideos['time'].dt.year
JPvideos['Year'] = JPvideos['time'].dt.year
```

```python
KRvideos.groupby("Year")["views"].mean()
```

```python
JPvideos.groupby("Year")["views"].mean()
```

```python
df_Korea = pd.DataFrame({'Year': ['2015','2017','2018'],
                  'meanviews_KR': [8720.7, 465636.5,411670.2]})
df_Korea
```

```python
df_Japan = pd.DataFrame({'Year': ['2015','2017','2018'],
                  'meanviews_JP': [6965, 6218,262071.3]})
df_Japan
```

```python
result = pd.merge(df_Korea, df_Japan, on = 'Year')
result
```

```python
result.plot.bar()
```

한국이 일본보다 더 높은 년도별 유튜브 평균 조회수를 기록하는 것을 알 수 있다.
